```r
library(Seurat)
library(cowplot)
library(tidyverse)
library(glue)
library(here)
```

```r
source(here::here("misc/paths.R"))
```

```r
se_obj <- "scRNAseq/Final_MSS_Pellka_all_cells.rds"%>%
  here::here() %>%
  glue::glue() %>%
  readRDS(.)
```

```r
colnames(se_obj@meta.data)
```

```r
options(repr.plot.width=15, repr.plot.height=10)  # Width = 10 inches, Height = 6 inches
```

```r
DimPlot(se_obj, group.by = c("annotation_lvl1"), cols = ggsci::pal_igv()(50),raster = F)
```

```r
DimPlot(se_obj, group.by = c("all_cells_lvl2"), cols = ggsci::pal_igv()(50),raster = F)
```

```r
Idents(se_obj) = se_obj$annotation_lvl1
all.markers <- FindAllMarkers(se_obj, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)
```

```r
"scRNAseq/MSS_Pellka_allcells_markers_lvl1.rds" %>%
  here::here() %>%
  glue::glue() %>%
  saveRDS(object = all.markers)

"{decon_sp}/{robj_dir}/mgs_lvl1.rds" %>%
    glue::glue() %>%
    here::here() %>%
    saveRDS(object = all.markers)
```

```r
Idents(se_obj) = se_obj$all_cells_lvl2
all.markers <- FindAllMarkers(se_obj, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)
```

```r
"scRNAseq/MSS_Pellka_allcells_markers_lvl2.rds" %>%
  here::here() %>%
  glue::glue() %>%
  saveRDS(object = all.markers)
```
