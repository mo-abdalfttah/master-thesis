```r
selected_res_minus_one <- 10
selected_res <- 11
clust.df <- "{hotspot_adata_tme}/{robj_dir}/clust_df.rds" %>%
  glue::glue() %>%
  here::here() %>%
  readRDS(file = .)
Modules = "{hotspot_adata_tme}/{robj_dir}/Modules_raw.rds" %>%
  glue::glue() %>%
  here::here() %>%
  readRDS(file = .)
```

```r
Metamodules <- list()
Metamodules_counts <- list()
clust.df[[selected_res_minus_one]]$Modules <- rownames(clust.df[[selected_res_minus_one]])

for(cluster in seq(1,selected_res)){
  curr.modules <- clust.df[[selected_res_minus_one]]$Modules[clust.df[[selected_res_minus_one]]$clust_res_11 == cluster]
  Metamodules <- append(Metamodules, list(unique(as.character(unlist(Modules[curr.modules], recursive = FALSE)))))
  Metamodules_counts <- append(Metamodules_counts, list(as.data.frame(table(as.character(unlist(Modules[curr.modules], recursive = FALSE))))))
}
names(Metamodules) <- paste0("TP", seq(1,selected_res))
names(Metamodules_counts) <- paste0("TP", seq(1,selected_res))
```

```r
all.genes <- c()

for (patient in patients) {
  file <- paste0("scRNAseq/TME/2-HotSpot/",patient,"/","Localcorrelations_",patient,"_30knn.csv")
  curr_pat <- as.data.frame(fread(file, header = TRUE))
  all.genes <- c(all.genes, curr_pat$V1)
  }

all.genes <- unique(all.genes)

"{hotspot_adata_tme}/{robj_dir}/GeneList_Correlations.csv" %>%
  glue::glue() %>%
  here::here() %>%
  write.csv(x = all.genes,file = .)
```

```r
#### Average correlation #####
all.genes <- "{hotspot_adata_tme}/{robj_dir}/GeneList_Correlations.csv" %>%
  glue::glue() %>%
  here::here() %>%
  read.csv(file = .)
all.genes <- all.genes$x

# Allocate final summatory matrixes
corr.mat <- matrix(0, length(all.genes), length(all.genes))
colnames(corr.mat) <- all.genes
rownames(corr.mat) <- all.genes
```

```r
# Correlation matrix per patient (curr.pat.m)
for (patient in patients) {
  print(patient)
  file <- paste0("scRNAseq/TME/2-HotSpot/",patient,"/","Localcorrelations_",patient,"_30knn.csv")
  curr.pat <- as.data.frame(fread(file, header = TRUE))
  rownames(curr.pat) <- curr.pat$V1
  curr.pat$V1 <- NULL
  curr.pat.m <- as.matrix(curr.pat)
  
  print("Correlation matrix opened")
  curr.pat.m <- as.matrix(curr.pat)
  
  print("Add current patient to summatorial matrix of correlations times weights")
  curr.mat <- matrix(0, ncol=length(all.genes), nrow=length(all.genes), dimnames=list(all.genes, all.genes))
  indxCurr <- outer(all.genes, all.genes, FUN=paste) %in% outer(rownames(curr.pat.m), colnames(curr.pat.m), FUN=paste)
  curr.mat[indxCurr] <- curr.pat.m
  corr.mat <- corr.mat + curr.mat
  
  print("Done!")
  gc()
}
```

```r
processTP <- function(tpName, avg.corr.mat, Metamodules, Metamodules_counts, clust.df) {
  TP <- avg.corr.mat[rownames(avg.corr.mat) %in% unlist(Metamodules[[tpName]]), colnames(avg.corr.mat) %in% unlist(Metamodules[[tpName]])]
  TP <- as.data.frame(rowMeans(TP))
  TP <- cbind(rownames(TP), TP, Metamodules_counts[[as.integer(sub("TP", "", tpName))]]$Freq[match(rownames(TP), Metamodules_counts[[as.integer(sub("TP", "", tpName))]]$Var1)])
  colnames(TP) <- c("Gene", "AverageCorrelation", "Counts")
  
  # Density Plot
  print(ggplot(TP) +
    geom_density(aes(y = AverageCorrelation)) +
    scale_x_discrete() +
    labs(title = tpName,
         y = "Average Correlation") +
    theme_classic())
  
  # Scatter Plot
  print(ggplot(TP, aes(x= AverageCorrelation, y= Counts)) +
    geom_point() +
    ggtitle(tpName) +
    geom_vline(xintercept = mean(TP$AverageCorrelation), linetype= "dashed", color= "red") +
    geom_hline(yintercept = length(unique(sub("_.*","",clust.df$Modules[clust.df$`10` == 1])))/2, linetype= "dashed", color= "blue") +
    theme_classic())
  
  # Filter and Rank
  TPf <- TP[(TP$AverageCorrelation >= mean(TP$AverageCorrelation) & TP$Counts >= length(unique(sub("_.*","",clust.df$Modules[clust.df$`10` == 1])))/2),]
  TPf$RAC <- rank(-TPf$AverageCorrelation, ties.method= "max")
  TPf$RCo <- rank(-TPf$Counts, ties.method= "first")
  
  # Rank Plot
  print(ggplot(TPf, aes(x= RAC, y= RCo)) +
    geom_point() +
    xlab("Average Correlation Rank") +
    ylab("Counts Rank") +
    ggtitle(tpName) +
    theme_classic())
  
  list(Genes = TPf$Gene[order(TPf$AverageCorrelation,decreasing = TRUE)])
}
```

```r
# Initialize TranscriptionPrograms list
TranscriptionPrograms <- list()

# Loop through TP1 to TPn
for(i in 1:selected_res) {
  tpName <- paste("TP", i, sep = "")
  TranscriptionPrograms[[tpName]] <- processTP(tpName, avg.corr.mat, Metamodules, Metamodules_counts, clust.df)$Genes
}

# Now TranscriptionPrograms list contains the results for all TPs
```

```r
"{hotspot_adata_tme}/{robj_dir}/TranscriptionPrograms_filtered_res_{selected_res}.rds" %>%
  glue::glue() %>%
  here::here() %>%
  saveRDS(object = TranscriptionPrograms,file = .)

"{hotspot_adata_tme}/{robj_dir}/TranscriptionPrograms_filtered_res_{selected_res}.xlsx" %>%
  glue::glue() %>%
  here::here() %>%
  write.xlsx(x = setNames(as.list(lapply(TranscriptionPrograms, data.frame)), names(TranscriptionPrograms)), file=.)
```
